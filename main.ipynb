```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
# abrir el archivo de inverviones csv y cargarlo a un dataframe
df = pd.read_csv('Archivo_de_Inversiones.csv')
df.head()
```

```python
def obtener_precio_google_finance(activo):
    try:
        # Construye la URL de búsqueda en Google Finance
        url = f"https://www.google.com/finance/quote/{activo}:NASDAQ"
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }
        
        # Realiza la solicitud HTTP
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"No se pudo obtener datos para {activo}")
            return None
        
        # Parsear el contenido HTML
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extraer el precio actual (puedes ajustar el selector según el sitio)
        precio = soup.find('div', class_='YMlKec fxKbKc').text.strip()
        return precio
    except Exception as e:
        print(f"Error al obtener el precio de {activo}: {e}")
        return None

# Añadir una columna con los precios
df['Precio_Actual'] = df['Activo'].apply(obtener_precio_google_finance)

df.head()
# # Guardar resultados en un archivo
# df.to_csv('precios_activos.csv', index=False)

# print(df)
```

```python
# convierte Precio_actual a int
df['Precio_Actual'] = df['Precio_Actual'].str.replace(',', '').str.replace('$', '').astype(float)
```

```python
# calculame la diferencia entre el precio de compra y el precio actual
df['Diferencia'] = df['Precio_Actual'] - df['Precio_Compra']
df.head()
```
